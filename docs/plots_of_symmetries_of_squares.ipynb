```python
x = np.linspace(-9, 9, 400)
y = np.linspace(-5, 5, 400)
x, y = np.meshgrid(x, y)
```

```python
def axes():
    plt.axhline(0, alpha=.1)
    plt.axvline(0, alpha=.1)
```

```python
plt.axhline(0)
plt.axvline(0)
a = 1.
b = 1.
plt.contour(x, y,(x**2/a**2 - y**2/b**2), [1], colors='b')
plt.contour(x, y,(-x**2/a**2 + y**2/b**2), [1], colors='b')
plt.contour(x, y,(+x**2/a**2 + y**2/b**2), [1], colors='k')
plt.contour(x, y,(2 * x * y), [1], colors='g')
plt.contour(x, y,(-2 * x * y), [1], colors='g')
#plt.contour(x, y,(x + y + 1), [1], colors='k')
#plt.contour(x, y,(x - y + 1), [1], colors='k')
plt.grid(True)
plt.axes().set_aspect('equal')
plt.xlim(-4, 4)
plt.ylim(-4, 4)

#plt.axline([0, 0], [1, 1])

#plt2.scatter(X, Y)
#plt2.plot([0, 1], [0, 1], color = 'black', linewidth = 2)

plt.plot([-4, 4], [-4, 4], ls="--", c=".3")
plt.plot([4, -4], [-4, 4], ls="--", c=".3")

plt.xticks(range(-3, 4), visible=False)

# plt.axes.get_xaxis().set_visible(False)

plt.tick_params(         # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    left=False,
    labelbottom=False)

plt.show();
```

```python
# Create a function that can draw this

def plot_square_symmetries(size: int=2, unity: bool=False, special_relativity: bool=False, gravity: bool=False, save: str=''):
    plt.axhline(0)
    plt.axvline(0)

    if unity:
        plt.contour(x, y,(+x**2 + y**2), [1], colors='k')
        
    if special_relativity:
        plt.contour(x, y,(x**2 - y**2), [1], colors='b')
        plt.contour(x, y,(-x**2 + y**2), [1], colors='b')
    
    if gravity:
        plt.contour(x, y,(2 * x * y), [1], colors='g')
        plt.contour(x, y,(-2 * x * y), [1], colors='g')

    plt.grid(True)
    plt.axes().set_aspect('equal')
    plt.xlim(-size, size)
    plt.ylim(-size, size)

    plt.plot([-size, size], [-size, size], ls="--", c=".3")
    plt.plot([size, -size], [-size, size], ls="--", c=".3")

    plt.xticks(range(-size, size), visible=False)
    plt.yticks(range(-size, size), visible=False)
    
    plt.tick_params(         # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    left=False,
    labelbottom=False)
    
    if save:
        name=save + ".png"
        plt.savefig(name, transparent=True)
        print(f"saved file: {name}")
    
    plt.show();
```

```python
plot_square_symmetries(size=2, unity=True, save="unity")
```

```python
plot_square_symmetries(special_relativity=True, save="special_relativity")
```

```python
plot_square_symmetries(gravity=True, save="gravity")
```
